# Audio Acquisition

```python
import sounddevice as sd
from scipy.io import wavfile as wav
import numpy as np
import os
import cv2 as cv
```

```python
rec_duration = 300
rec_rate = 44100
i=1
member='Alberto'
```

```python
print('La registrazione numero {} di {} è iniziata'.format(i,member))
record = sd.rec(int(rec_duration * rec_rate), samplerate=rec_rate, channels=1, blocking=True)
print('La registrazione numero {} di {} è terminata'.format(i,member))
wav.write('audio/rec{}{}.wav'.format(member,i), rec_rate, record)
```

```python
def audio_splitter(audio_path, seconds, member):
    tracklist = []
    audio_rate, audio = wav.read(audio_path)
    for i in np.arange(int((len(audio)/audio_rate)/seconds)):
        track = audio[(audio_rate*i*seconds):(audio_rate*(i+1)*seconds)]
        wav.write('dataset/{}_{}.wav'.format(member,i), rate=audio_rate, data=track)
```

```python
#Audio Acquisition
seconds=2
audio_splitter('audio/recAlberto1.wav', seconds, 'alberto_1')
audio_splitter('audio/recAlberto2.wav', seconds, 'alberto_2')
audio_splitter('audio/recAlberto3.wav', seconds, 'alberto_3')
audio_splitter('audio/recAlice1.wav', seconds, 'alice_1')
audio_splitter('audio/recAlice2.wav', seconds, 'alice_2')
audio_splitter('audio/recAlice3.wav', seconds, 'alice_3')
audio_splitter('audio/recDavide1.wav', seconds, 'davide_1')
audio_splitter('audio/recDavide2.wav', seconds, 'davide_2')
audio_splitter('audio/recDavide3.wav', seconds, 'davide_3')
```

# Image Acquisition

```python
face_detector = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
def image_acquisition(path, num_images, member):
    for i in range(num_images):
        cap = cv.VideoCapture(0)
        result, img = cap.read()

        faces = face_detector.detectMultiScale(img, minNeighbors=10, minSize=(10,10))
        
        print("[INFO] Found {0} Faces!".format(len(faces)))

        for (x,y,w,h) in faces:
            face = img[y:y+h,x:x+h,:]
            cv.imwrite(path+'/{}{}.jpg'.format(member, i), face)

    cap.release()
```
