```python
from kafka import KafkaConsumer
import json
from pprint import pprint
```

```python
con1 = KafkaConsumer( bootstrap_servers=["kafka:9092"],auto_offset_reset="latest",
                     value_deserializer=lambda v: json.loads(v.decode("utf-8")))

con1.subscribe(['nazioni'])                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                  
```

```python

with open("nazioni.json", "a") as file:
    for msg in con1:
        data = msg.value
        json.dump(data, file)
        file.write("\n")
```
